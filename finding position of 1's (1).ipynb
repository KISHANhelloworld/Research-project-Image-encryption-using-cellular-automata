```python

def bits1(n):
    b = []
    
    
    while n:
        b = [n & 1] + b
        n >>= 1
    return b or [0]
for i in range(255):
    #print(i,end='---->')            ## print decimal no
    #print (bits1(i),end='--->')     ## binary of same number
    c=bits1(i)                      ##storing list in to variable 
    ##finding position of 1's
    
    for ele in range(len(c)):        
        if c[ele]==1:
            print(ele,end=',')         #position of one
    print()                         # giving gap      
    
    
```
