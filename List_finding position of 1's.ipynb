```python

def bits1(n):
    b = []
    
    
    while n:
        b = [n & 1] + b
        n >>= 1
    return b or [0]
for i in range(1,256):
   # print(i,end=',')            ## print decimal no
   # print (bits1(i),end=',')     ## binary of same number
    c=bits1(i)                      ##storing list in to variable 
   
    list1=[]                   ## creat empty list to store 1's index.
    ##finding position of 1's
    for ele in range(len(c)):  
        
        if c[ele]==1:
            list1.append(ele)   # append element in list1[].
    print(list1)        
            
                    
                          # giving gap      
    
    
```
